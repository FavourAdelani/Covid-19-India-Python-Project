 This notebook performs an exploratory data analysis on COVID-19 Data for India gotten from KAGGLE, it data contains information relevant to the corona virus pandemic.
 This notebook  uses pandas and matplotlib libraries to extract insights, visualize data and generate meaningful graphs and statistics.
 Each section focuses on specific aspects of COVID-19 in India statistics and provide valuable insights.
 This  analysis aims to provide a comprehensive overview of this dataset and various influential factors to consider.

```python
#import necessary libraries
import pandas as pd
import numpy as np
import os 
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from plotly.subplots import make_subplots
from datetime import datetime
```

```python
# dropping columns that are not relevant to our analysis from the covid dataframe
covid.drop(["Sno","Time","ConfirmedIndianNational","ConfirmedForeignNational"], inplace = True, axis =1)
```

```python
#Viewing the updated covid dataframe
covid.head()
```

```python
#changing the format for the date column
covid['Date'] = pd.to_datetime(covid['Date'], format ='%Y-%m-%d')
```

```python
#viewing the updated covid dataframe
covid.head()
```

```python
#finding active cases
covid['Active Cases'] = covid['Confirmed'] - (covid['Cured'] + covid['Deaths']) 
covid.tail()
```

```python
#creating a pivot table
statewise = pd.pivot_table(covid, values = ["Confirmed", "Deaths", "Cured"], index = "State/UnionTerritory", aggfunc = max)
```

```python
#finding recovery rate
statewise["Recovery Rate"] = statewise["Cured"]*100/ statewise["Confirmed"]
```

```python
#finding mortality rate
statewise["Mortality Rate"] = statewise["Deaths"]*100/ statewise["Confirmed"]
```

```python
#sorting the statewise column using confirmed cases in descending order
statewise =statewise.sort_values(by = "Confirmed", ascending = False)
```

```python
#creating a visual for the statewise pivot table
statewise.style.background_gradient(cmap= "cubehelix")
```

```python
#Top 10 active cases states

top_10_active_cases = covid.groupby(by = 'State/UnionTerritory').max() [['Active Cases', 'Date']].sort_values (by = ['Active Cases'], ascending =False).reset_index()
```

```python
#plotting a figure for our chart

fig = plt.figure(figsize=(16,9))
```

```python
#plotting a bar chart to visualize the top 10 active cases dataframe, we start with the title

plt.title("Top 10 states with the most active cases in India", size= 25)
```

```python
#defining the x and y axis of the chart

ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active Cases", x ="State/UnionTerritory", linewidth = 2, edgecolor ='green')
```

```python
#Top 10 active cases states

top_10_active_cases = covid.groupby(by = 'State/UnionTerritory').max() [['Active Cases', 'Date']].sort_values (by = ['Active Cases'], ascending =False).reset_index()

#plotting a figure for our chart

fig = plt.figure(figsize=(16,9))

#plotting a bar chart to visualize the top 10 active cases dataframe, we start with the title

plt.title("Top 10 states with the most active cases in India", size= 25)

#defining the x and y axis of the chart

ax = sns.barplot(data = top_10_active_cases.iloc[:10], y = "Active Cases", x ="State/UnionTerritory", linewidth = 2, edgecolor ='green')

#adjusting the overlapping labels
plt.xlabel ("States")
plt.ylabel ("Total Active Cases")
plt.show()
```

```python
#Top 10 states with the highest deaths

top_10_deaths = covid.groupby(by = 'State/UnionTerritory').max() [['Deaths', 'Date']].sort_values (by = ['Deaths'], ascending =False).reset_index()

fig = plt.figure(figsize= (18,5))

plt.title("Top 10 states with the most deaths in India", size = 25)


ax = sns.barplot(data = top_10_deaths.iloc[:12], y = "Deaths", x ="State/UnionTerritory", linewidth = 2, edgecolor ='red')

plt.xlabel ("States")
plt.ylabel ("Total Death Cases")
plt.show()
```

```python
#we want to know the total number of missing values in the dataframe

vaccine.isnull().sum()
```

```python
#dropping missing value columns
vaccination = vaccine.drop(columns= ['Sputnik V (Doses Administered)', 'AEFI', '18-44 Years (Doses Administered)', '45-60 Years (Doses Administered)', '60+ Years (Doses Administered)'], axis =1)
```

```python
#now to check the updated vaccination dataframe

vaccination.head()
```

```python
# Male vs Female vaccination Pie plot

male= vaccination["Male(Individuals Vaccinated)"].sum()
female= vaccination["Female(Individuals Vaccinated)"].sum()
px.pie(names= ["Male", "Female"], values= [male, female], title= "Male and Female Vaccination")
```

```python
# Return states not = India

vaccine_not =vaccine[vaccine.State!= 'India']
vaccine_not
```

```python
#Renaming the total individuals vaccinated column to just total

vaccine_not.rename(columns= {"Total Individuals Vaccinated" : "Total"}, inplace= True)
vaccine_not.head()
```

```python
# Most Vaccinated States

max_vac= vaccine_not.groupby ('State') ['Total'].sum().to_frame('Total')
max_vac= max_vac.sort_values ('Total', ascending= False)[:5]
max_vac
```

```python
#plotting a bar chart for the data in max_vac

fig = plt.figure(figsize= (10,5))
plt.title ("Top 5 Vaccinated States In India", size= 20)
x = sns.barplot(data= max_vac.iloc[:10], y = max_vac.Total, x= max_vac.index, linewidth= 2, edgecolor = 'black')
plt.xlabel ("States")
plt.ylabel ("Vaccination")
plt.show()
```

```python
# least Vaccinated States

min_vac= vaccine_not.groupby ('State') ['Total'].sum().to_frame('Total')
min_vac= max_vac.sort_values ('Total', ascending= True)[:5]
min_vac
```

```python
#plotting a bar graph for the least vaccinated states

fig = plt.figure(figsize= (10,5))
plt.title ("Least 5 Vaccinated States In India", size= 20)
x = sns.barplot(data= min_vac.iloc[:10], y = min_vac.Total, x= min_vac.index, linewidth= 2, edgecolor = 'black')
plt.xlabel ("States")
plt.ylabel ("Vaccination")
plt.show()
```
